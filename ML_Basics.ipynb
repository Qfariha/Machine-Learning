# **Task-1**

```python
#Task-1
import numpy as np

#1
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

#2
a = a.reshape(5, 4)
b = b.reshape(5, 4)

#3
result= a*b
print(result)

#4
#print(b)
b[0][1]=15
#print(b)
c = a @ b.T
print(c)

#5
print("Columns 2 and 3 of matrix c:\n", c[:, 1:3])
print("Whole matrix c:\n", c[:, :])

#6
largest_val_index = np.unravel_index(np.argmax(c), c.shape)
#np.argmax(c)=>flat index of the maximum element
#np.unravel_index=> maps flat index to multi-dimensional coordinates
print("Largest Element: ", c[largest_val_index])
print("Largest Element Index: ", largest_val_index)

#7
c_flattened = c.flatten()
print("1-D array:\n", c_flattened)
```
